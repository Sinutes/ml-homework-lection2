Дан файл "athlete_events", который содержит информацию об олимпийских чемпионах за последние 120 лет.

```python
%matplotlib inline
import zipfile
import numpy as np
import pandas as pd
import seaborn as sns

z = zipfile.ZipFile("../data/athlete_events.zip")
df = pd.read_csv(z.open("athlete_events.csv"))
df = df.dropna(subset=['Medal', "Age", "Height", "Weight"])
df.head()
```

**1. Сколько мужчин и женщин получили золотые, серебрянные и бронзовые медали?**

```python
df['Medal'].value_counts()
```

**2. Какая страна получила наибольшее количество золотых медалей за всю историю олимпийских игр?**

```python
golden_medalyst_df = df.loc[df['Medal'] == 'Gold']
golden_medalyst_df['Team'].value_counts()[0:1]
```

**3. Выведите распределение пола участника олимпиады от вида спорта (crosstab)**

```python
pd.crosstab(df['Sex'], df['Sport'], margins=True)
```

**4. Выведите средний возраст и его стандартное отклонения для женщин, учавствовавших в хоккее на льду**

```python
female_hockey_df = df.loc[(df['Sport'] == 'Ice Hockey') & (df['Sex'] == 'F')]
print('Средний возраст женщин, учавствовавших в хоккее на льду:', female_hockey_df['Age'].mean())
print('Стандартное отклонение: ', female_hockey_df['Age'].std())
```

**5. У какой страны больше всего было больше всего женщин, получивших бронзовую медаль?**

```python
bronze_female_df = df.loc[(df['Medal'] == 'Bronze') & (df['Sex'] == 'F')]
bronze_female_df['Team'].value_counts()[0:1]
```

**6. Постройте гистограмму распределения количества медалей (бронза, серебро, золото) для первых трех стран, получивших наибольшее количество медалей**

```python
three_most_medal_countries = df['Team'].value_counts()[0:3]
print('3 страны с наибольшим количеством медалей:\n', three_most_medal_countries, '\n')

df_data = []
for country in three_most_medal_countries.index.values:
    data = df.loc[df['Team'] == country]['Medal'].value_counts()
    df_data.append([country, data[0], data[1], data[2]])

three_most_medal_countries_df = pd.DataFrame(df_data, columns = ['Country', 'Gold', 'Silver', 'Bronze'])
print(three_most_medal_countries_df)

sns.distplot( three_most_medal_countries_df["Gold"] , color="red")
sns.distplot( three_most_medal_countries_df["Silver"] , color="blue")
sns.distplot( three_most_medal_countries_df["Bronze"] , color="green")
```

**7. Нарисуйте распределение веса мужчин, получивших серебрянную медаль(density или distplot)**

```python
male_silver_df = df.loc[(df['Sex'] == 'M') & (df['Medal'] == 'Silver')]
sns.distplot(male_silver_df['Weight'] , color='red')
```

**8. Постройте boxplot для возраста участника в зависимости от медали**

```python
sns.boxplot(y='Age', x='Medal', data=df)
```

**9. Постройте pairplot для веса, возраста и роста участников от USA.**
